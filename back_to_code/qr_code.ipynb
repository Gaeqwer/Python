```python
import qrcode

data = "https://live.staticflickr.com/6068/6065510171_bb00e7222b_z.jpg" #the thing you want to show as qrcode.

qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)

qr.add_data(data)
qr.make(fit=True)

img = qr.make_image(fill_color="black",back_color="white")

img.save("my_qrcode.png")

print("Success")
```
