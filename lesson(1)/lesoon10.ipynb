```python
import pandas as pd
scores = [
    {'Chinese':97,'English':100,'Math':94,'History':90,'Discover':99},
    {'Chinese':91,'English':90,'Math':84,'History':67,'Discover':69}
]
dataFrame = pd.DataFrame(index=["Gary","Cody"],data=scores)
display(dataFrame)
```

```python
subjects = list(dataFrame.columns)
subjects
```

```python
swapdataFrame = dataFrame.T
display(swapdataFrame)
Gary_se = swapdataFrame['Gary']
Cody_se = swapdataFrame['Cody']
```

```python
Gary_sc = Gary_se.to_list()
Cody_sc = Cody_se.to_list()
display(Gary_sc)
display(Cody_sc)
```

```python
import matplotlib.pyplot as plt
figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,Gary_sc,'bo-',label='Gary')
axes.plot(subjects,Cody_sc,'ro-',label='Cody')
axes.legend()
plt.show()
```

```python
import pandas as pd
scores = [
    {'Chinese':97,'English':100,'Math':94,'History':90,'Discover':99},
    {'Chinese':91,'English':90,'Math':84,'History':67,'Discover':69}
]
dataFrame = pd.DataFrame(index=["Gary","Cody"],data=scores)
display(dataFrame.T)
```

```python
import matplotlib.pyplot as plt
figure = plt.figure()
axes = figure.add_subplot(1,1,1)
dataFrame.T.plot(ax=axes,style=['--','-'],marker='o')
axes.set_title('Scores')
axes.grid(visible=True,axis='y')
plt.show()
```
