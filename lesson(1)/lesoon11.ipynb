```python
import numpy as np
import pandas as pd
scores = np.random.randint(50,101,size=(50,5))
scores_pd = pd.DataFrame(scores,
             columns=['Chinese','English','Math','Geogrophy','History'],
             index = range(1,51)
             )
scores_pd
```

```python
scores_pd.index.name = "Number"
scores_pd.columns.name = "Subject"
scores_pd
```

```python
display(scores_pd['Chinese'][:10])
scores_pd['Math'][:10]
```

```python
scores_pd['Chinese'] + scores_pd['English'] + scores_pd['Geogrophy'] + scores_pd['History'] + scores_pd['Math']
```

```python
scores_pd['Total'] = scores_pd[['Chinese','English','Math','Geogrophy','History']].sum(axis=1)
scores_pd
scores_pd['Average'] = scores_pd[['Chinese','English','Math','Geogrophy','History']].mean(axis=1)
scores_pd
```

```python
scores_pd['Rank'] = scores_pd['Total'].rank(method='first',ascending=False)
scores_pd
```

```python
scores_pd.sort_values(by='Rank')
```

```python
def format60(v:int):
    if v >= 60:
        return 'Pass'
    else:
        return 'Not pass'
scores_pd[['Chinese','English','Math','Geogrophy','History']].style.format(format60)
```

```python
scores_pd[['Chinese','English','Math','Geogrophy','History']].style.highlight_between(
left=0,right=59,color='blue')
```
